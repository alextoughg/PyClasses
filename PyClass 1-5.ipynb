## Assignment operator

Most of us deal with a large amounts of emails each day, and sometimes our names are just too long to be put into every single email we send out. So we just use our initials, which is like a **placeholder** for our name. 

```python
AT = 'Alex Tough'
```

This is our first encounter with the assignment operator. Intuitively, the **assignment operator** attaches a name or a **placeholder** to a thing, much like our initials are a placeholder for our name. The statement 

```python
AT = 'Alex Tough'
```

is read as "the variable **AT** has value "Alex Tough"". Whatever is to the left of the equals (=) sign is called a **variable** and whatever is to the right of the equals sign is called a **value**. 

Another helpful way to visualize the assignment operator is as a **container**: the variable is the label of the container, and the value is the content of the container.

## Values (types) 

We are now clear on what can go on the left hand side of an assignment. This is, we know what type of labels we can put on containers. Now let's see what can go on the right hand of an assignment, this is, what values can we put in our containers. 

For text processing tasks, the most common values (or **types**) we can put into our containers are:
1. Strings (a sequence of characters, as in 'A Hundred Years of Solitude)

3. Lists (I'll explain this in a bit) 

Let's take a brief look at each type. 

### Strings

```python
hundred_years_single
```

```python
hundred_years_double
```

### Lists

Recall from a previous exercise our variable with the list of novel names:

```python
MC = 'Love in the Time of Cholera'
YI = '100 Years of Solitude'
SS = 'Chronicle of a Death Foretold'
```

Say we wanted to count the number of characters in each title. We could then do

There is nothing wrong with this approach, but wouldn't it be easier to do

We can group these titles by using a **list**. In Python, lists are created by comma-separated values in square brackets.  

```python
title_list = ['Love in the Time of Cholera', '100 Years of Solitude', 'Chronicle of a Death Foretold']
title_list
```

So our previous task is now simplified to

```python
count characters of title_list
```

Note that we could also build the list using variables instead of values 

```python
title_list = [MC, YI, SS]
title_list
```

## For-Loops

Let's recall our previous task. 

```python
count characters of title_list
```

Now how do we actually go through this list? We can use a **for loop**:

```python
for title in title_list:
    count characters of title
```

The above code says: for each title in title_list, count the characters of each title. Therefore a for-loop allows us to apply an action to each element of a list. Notice that the indentation of the code after the for-loop was no accident: it is the only way Python knows which code will be applied to each element in the loop. In our example, only "count characters of title" will be applied to each element in the loop as it is the only code that is indented.

For example, let's say I wanted to find out the length of the list title_list. How could I accomplish this using a for-loop? 

One way to do this is to count the elements of the list. What tools do we have to count that we have learned so far? Our **counter variables**! 

```python
count = 0
count = count + 1
```

Every time we see an element in the list we can increment our count variable by 1. In code:

```python
title_list_count = 0
for title in title_list:
    title_list_count = title_list_count + 1
title_list_count
```

Notice here that once we write code that has the same indentation as the word 'for', our loop is over.

## I/O (Input/Output)

Notice that up to this point, to see the contents of a variable, we have been putting the variable at the end of each fragment of code 

```python
title_list_count = 0
for title in title_list:
    title_list_count = title_list_count + 1
title_list_count
```

However, this doesn't seem very intuitive. Can't we just say something like "print title_list_count"? In fact, we can using Python's **print** command. 

```python
title_list_count = 0
for title in title_list:
    title_list_count = title_list_count + 1
print(title_list_count)
```

The print command is very convenient. It allows us to print multiple things in one line.  

```python
print("The length of the list is", title_list_count)
```
